```python
from functools import reduce


data_frames = [df1704, df1708, df1711, df1810, df1904,df1907]
master_abundance=reduce(lambda  left,right: pd.merge(left,right,on=['genus'], how='outer'), data_frames)

master_abundance.to_csv('All_stains_fresh_genus_counts.csv')

master_abundance=master_abundance.fillna(0)
master_abundance
```

```python
fold=master_abundance[['genus']].copy()

fold['170412_fold_change']=master_abundance['170412_RSG']/master_abundance['170412_Syto']
fold['170818_fold_change']=master_abundance['170818_RSG']/master_abundance['170818_Syto']
fold['171102_fold_change']=master_abundance['171102_RSG']/master_abundance['171102_Syto']
fold['181030_fold_change']=master_abundance['181030_RSG']/master_abundance['181030_Syto']
fold['190402_fold_change']=master_abundance['190402_RSG']/master_abundance['190402_Syto']
fold['190709_fold_change']=master_abundance['190709_RSG']/master_abundance['190709_Syto']

#Notes

#Possible options
#present in both >> color scale (This will be a number)
#present in RSG but not Syto9 >> green (this is represented as "inf") change to appear
#present in Syto9 but not RSG >>red (this is 0.0) change to disappera
#present in neither blank. >> this should be white this is nan

fold=fold.set_index('genus')
fold.replace(to_replace=0, value='Disappear', inplace=True)
fold.replace(to_replace=np.inf, value='Appear', inplace=True)
fold.dropna(how='all', inplace=True, axis=0)
fold.to_csv(outdir+ 'fold_change_Syto9_to_RSG_stain.csv')
```
