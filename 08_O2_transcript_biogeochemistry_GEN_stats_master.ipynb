```python
import pandas as pd
from pandas import DataFrame
import os
import sys
import csv
from pathlib import Path
import seaborn as sns
sns.set()
import numpy as np
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
plt.show()
from collections import defaultdict
import os.path as op
import glob

os.chdir('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/')
outdir='/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/O2_consumption/'
matplotlib.__version__
```

```python

transcript_data=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/O2_consumption/All_GoM_RNA_seq_transcripts_per_cell_by_genus_and_function_stacked_weighted_avg_O2_consumed_ALL_genes_all_Genus.csv',
                            index_col=0)

transcript_data=transcript_data.drop_duplicates()

transcript_pivot=transcript_data.pivot(index=['genus','date'], values='transcripts_per_cell', columns='ko')
transcript_pivot.to_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/GORG_recruitment/RNA_summaries/All_GoM_RNA_seq_transcripts_per_cell_by_genus_and_KO.csv')

transcript_pivot
#transcript_data
```

```python
O2_data=transcript_data[['genus', 'date', 'Weighted_avg_O2_consumed_per_cell']]
O2_data=O2_data.drop_duplicates()

O2_data
```

```python
master_data=transcript_pivot.merge(O2_data, on=['genus', 'date'])

master_data
```

```python
columns=['name', 'final_assembly_length', 'gc_content', 'estimated_diameter', 'Sample_collection_Date', 'gate',
         'GTDB_classification', 'checkM_estimated_completeness', 'total_CDS_count', 'hypothetical_cds_fraction']

SAG_data=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/All_GoM_SAGs_1cell_20kb_decon_531normalized_predresp_rate_GTDBclass.csv', 
                     usecols=columns)
SAG_data.head()

SAG_data['total_estimated_genome_length']=SAG_data['final_assembly_length']/(SAG_data['checkM_estimated_completeness']/100)

SAG_data['total_estimated_genome_length']=SAG_data['total_estimated_genome_length'].replace(np.inf, np.nan)
SAG_data
```

```python
GEN_pivot=pd.pivot_table(SAG_data, index='GTDB_classification', aggfunc=np.mean, values=['gc_content', 'hypothetical_cds_fraction',])
GEN_pivot=GEN_pivot.rename(columns={'gc_content':'Average_gc_content',
                                   'hypothetical_cds_fraction':'Average_hypothetical_cds_fraction'})
GEN_pivot
```

```python
size_pivot=pd.pivot_table(SAG_data, index='GTDB_classification', aggfunc=np.median, values='estimated_diameter')
size_pivot=size_pivot.rename(columns={'estimated_diameter':'median_estimated_diameter'})
size_pivot
```

```python
GEN_pivot=GEN_pivot.merge(size_pivot, how='left', on='GTDB_classification')
GEN_pivot
```

```python
fiftyperc_GENs=SAG_data.loc[SAG_data['checkM_estimated_completeness'] >= 50]
fiftyperc_GENs
fiftyperc_GEN_pivot=pd.pivot_table(SAG_data, index='GTDB_classification', aggfunc=np.mean, values='total_estimated_genome_length')
fiftyperc_GEN_pivot=fiftyperc_GEN_pivot.rename(columns={'total_estimated_genome_length':'mean_estimated_genome_length'})
fiftyperc_GEN_pivot
```

```python
GEN_pivot=GEN_pivot.merge(fiftyperc_GEN_pivot, how='left', on='GTDB_classification')
GEN_pivot
```

```python
rates=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/gRodon/genus_doubling_rates.csv', index_col=0)
rates
```

```python

GEN_stats=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/GTDB_GoM_SAG_classification_key.csv')
GEN_stats=GEN_stats.merge(GEN_pivot, left_on='GTDB_genus', right_on='GTDB_classification')
GEN_stats=GEN_stats.merge(rates, left_on='GTDB_genus', right_on='genus')
GEN_stats=GEN_stats.drop(columns={'genus'})

GEN_stats.to_csv(outdir+ 'Genus_statistics.csv')
GEN_stats.to_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/Genus_statistics.csv')
GEN_stats
```
