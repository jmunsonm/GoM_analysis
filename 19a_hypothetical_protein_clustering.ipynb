```python
# Added by Jacob on April 29th 2021 To summarize genes by genus and prokka annotations.
# I added this because after looking over the KO database I realized that proteorhodopsins are very poorly 
# characterized in the database and that if I am interested in those genes I should probably use a different 
# annotation methodology for these

# counting exclusive and shared reads by genus... similar to loop at the end of above cell, except grouping by genus and prokka annotations instead of by genus and ko

final_outdir = safe_makedir("/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/GORG_recruitment_mmseq_clusts3/summaries")
outfiles = glob.glob('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/GORG_recruitment_mmseq_clusts3/All*/') + glob.glob('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/GORG_recruitment_mmseq_clusts3/ALL*/')

for o in outfiles:
    dflist = []
    csvs = glob.glob(op.join(o, '*.tsv'))
    for c in csvs:
        sdf = pd.read_csv(c, sep="\t", engine='python')
        
        sgdf = sdf.groupby(['genus','gene_clusters'])['exclusive','shared'].sum()
        dflist.append(sgdf)

    fdf = pd.concat(dflist).reset_index()
    fdf.to_csv(op.join(final_outdir, '{}_read_count_by_genus_and_prokka_annotations.csv'.format(o.split("/")[-2])), index=False)
```
