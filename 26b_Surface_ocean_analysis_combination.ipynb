```python
# remove SAGs that have been identified as contaminated via tetramer analysis 
# find all the SSU fasta files, grep the file names using a txt file, delete the ones with a partial match in the txt file
!find Surface_SSU_seqs/* | grep -f /mnt/scgc/simon/microg2p/Data/Lists/contaminated_SAGs.txt | xargs rm

#create a fasta file for every genus
outdir='/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/Surface_ocean_SAG_analysis/Surface_SSU_seqs/'
!rm *.fasta
SSU_dir=glob.glob('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/Surface_ocean_SAG_analysis/Surface_SSU_seqs/*.fasta')

for genus in tenSAG:
    df=SAG_data[SAG_data['GTDB_classification'] == genus]
    SAGs=df['name'].to_list()

    with open(genus+'_SSU_sequences.fa', 'w') as out:
        for i in SAGs:
            for infile in SSU_dir:
                if i in infile:
                    with open(infile, 'r') as infile:
                        out.write(infile.read())
                        
!mv *.fa Surface_SAGs_SSU_by_genus/
```

```python
os.chdir('Surface_SAGs_SSU_by_genus/')
!pwd
# sort the 16S sequences long to short
!for file in *_SSU_sequences.fa; do seqkit sort $file -o ${file/%SSU_sequences.fa/sorted.fa} -l -r; done

# extract just the longest 16S sequence
!for file in *sorted.fa; do awk '/^>/{if(N)exit;++N;} {print;}' $file > ${file/%sorted.fa/longest16S.fa}; done

#remove the sorted files
!rm *sorted.fa
```

```python
# rename the name of the longest sequence so that it is the name of the genus
# This helps with future annotations of a tree where I have to refer to the branch length
fasta_files=glob.glob('*16S.fa')

for i in fasta_files:
    
    genus=i.split("_lon")[0]
    genus=str(genus)
   # print(genus)
    fasta_in = open(i)

    # Create FASTA output file.
    fasta_out = open(genus+'_16Srep.fa', 'w')

    for line in fasta_in.readlines():
        if line.startswith('>'):
            contig_id = '>' + str(genus) + '\n'
 #           print(contig_id)
            fasta_out.write(contig_id)
        else:
            fasta_out.write(line)
    fasta_out.close()
    fasta_in.close()
```
