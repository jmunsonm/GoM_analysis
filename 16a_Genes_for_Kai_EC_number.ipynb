```python
import pandas as pd
from pandas import DataFrame
import os
import sys
import csv
from pathlib import Path
import seaborn as sns
sns.set()
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.show()
from collections import defaultdict
import os.path as op
import glob

os.chdir('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/For_kai')
outdir=('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/For_kai/')
matplotlib.__version__
```

```python
transcripts=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/GORG_recruitment/RNA_summaries/All_GoM_RNA_seq_transcripts_per_cell_by_genus_and_prokka_annotation_pivot.csv',
                          index_col=[0,1])

transcripts
```

```python
KO_transcripts=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/Transcripts_percell_O2_percell_Genus_stats_biogeochemical_master_no2017.csv',
                          index_col=[1,2])
tax=['full_GTDBtk_classification']

classification=KO_transcripts.filter(tax)
classification
```

```python

KO_names=transcripts.columns.values.tolist()
KO_names

gluc_num='3.2.1.21'
xylan_num='3.2.1.37'

gluc_names=[]
xylan_names=[]

for i in KO_names:
    if gluc_num in i:
        gluc_names.append(i)
    if xylan_num in i:
        xylan_names.append(i)


print(xylan_names)
```

```python


glucosidases_KO=transcripts.filter(gluc_names)
glucosidases_KO=glucosidases_KO.dropna(how='all',axis=0)
glucosidases_KO=glucosidases_KO.merge(classification, on=['genus', 'date'], how='left')

glucosidases_KO.to_csv(outdir+'betaglucoside_EC_transcripts_per_cell.csv')
glucosidases_KO
```

```python

xylan_KO=transcripts.filter(xylan_names)
xylan_KO=xylan_KO.dropna(how='all',axis=0)
xylan_KO=xylan_KO.merge(classification, on=['genus', 'date'], how='left')

xylan_KO.to_csv(outdir+'Xylanases_EC_transcripts_per_cell.csv')
xylan_KO
```
