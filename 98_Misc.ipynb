```python
Total=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/O2_consumption/O2_consumed_by_cells_above_and_below_detection_limit_per_genus.csv',
                 index_col=0)
Total=Total[Total['date']!= 20171102]
Total=Total[Total['cells/ml']!=0]
Total=Total.drop('fraction', axis=1)
Total=Total.groupby(['genus', 'date']).sum()
Total.to_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/20210325_GoM_recluster_analysis/O2_consumption/Total_O2_consumption_and_abundance_per_genus.csv')
Total=Total.reset_index()
Total=Total.merge(family, on='genus', how='left')
Total=Total.rename(columns={'fmol_O2_consumed_per_ml_per_hr':'total_fmol_O2_consumed_per_ml_per_hr'})
Total
```

```python
Total_2018=Total[Total['date']==20181030]
Total_2018=Total_2018.sort_values(by='total_fmol_O2_consumed_per_ml_per_hr', ascending=False)
Last=Total_2018[Total_2018['genus'] == 'Unclassified']
Total_2018=Total_2018[Total_2018['genus']!= 'Unclassified']
Total_2018=Total_2018.append(Last)
Total_2018['O2_perc']=(Total_2018['total_fmol_O2_consumed_per_ml_per_hr']/Total_2018['total_fmol_O2_consumed_per_ml_per_hr'].sum())*100
Total_2018['cells_perc']=(Total_2018['cells/ml']/Total_2018['cells/ml'].sum())*100

Rhod_2018=Total_2018[Total_2018['Family']== 'Rhodobacteraceae']
print(Rhod_2018[['O2_perc', 'cells_perc']].sum())

Total_2018['cumultative_cellperc']=Total_2018['cells_perc'].cumsum()
Total_2018['cumultative_O2_consumed']=Total_2018['O2_perc'].cumsum()

Total_2018
```
