```python
contig_files=glob.glob('/mnt/scgc/simon/microg2p/Data/SAG_contigs/cryo_RSG_contigs/*.fasta')+\
glob.glob('/mnt/scgc/simon/microg2p/Data/SAG_contigs/2021-08_SAG_contigs/*.fasta')+\
glob.glob('/mnt/scgc/simon/microg2p/Data/SAG_contigs/GoM/1*/*.fasta')+\
glob.glob('/mnt/scgc/simon/microg2p/Data/SAG_contigs/Poseidon/*.fasta')+\
glob.glob('/mnt/scgc/simon/microg2p/Data/SAG_contigs/Radprof/*.fasta')+\
glob.glob('/mnt/scgc/simon/microg2p/Data/SAG_contigs/AT42-11/*.fasta')+\
glob.glob('/mnt/scgc/simon/microg2p/Data/Herndl_WGS/contigs/AH-716*.fasta')
print(len(contig_files))

cp_files=[]

for i in SAGs:
    for file in contig_files:
        if i in file:
            cp_files.append(file)
print(len(cp_files))   


cp_files=list(set(cp_files))
len(cp_files)

```
