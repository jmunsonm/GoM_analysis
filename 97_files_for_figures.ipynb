```python
# FACS dot plots


cols=['name', 'Plate', 'Predicted_respiration_rate_fmol_O2perhr', 'GTDB_classification', 'gate', 'FACS_date', 'Sample_collection_Date', 'GTDBtk_classification']
SAGs=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/All_GoM_SAGs_1cell_20kb_decon_531normalized_predresp_rate_GTDBclass.csv',
                usecols=cols)

SAGs=SAGs[SAGs['FACS_date']==SAGs['Sample_collection_Date']]
SAGs=SAGs[SAGs['gate']=='Respiring cells (RSG-positive)']
SAGs[['domain', 'phyla', 'class', 'order', 'family', 'genus', 'species']]=SAGs.GTDBtk_classification.str.split(';', expand=True)

FACS_dot_plots=SAGs[['name', 'class', 'Plate']]

FACS_dot_plots['class']=FACS_dot_plots['class'].str[3:]

Classes=['Alphaproteobacteria', 'Gammaproteobacteria', 'Planctomycetes', 'Poseidoniia', 'Nitrososphaeria', 'Verrucomicrobiae', 'Acidimicrobiia',
         'Actinomycetia', 'Marinisomatia', 'Kapabacteria', 'Bacteroidia', 'Cyanobacteriia', 'UBA1135', 'Campylobacteria']

FACS_dot_plots['group']=np.where(FACS_dot_plots['class'].isin(Classes), FACS_dot_plots['class'], 'Other')
FACS_dot_plots

plates=FACS_dot_plots['Plate'].to_list()
plates=list(set(plates))
plates
for i in plates:
    df=FACS_dot_plots.loc[FACS_dot_plots['Plate'] == i ]
    df=df.drop(columns=['class', 'Plate'])
    i=str(i)
    df.to_csv(outdir+i+'_file_for_FACS_plot.csv', index=False)

df
```
