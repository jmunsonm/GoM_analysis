```python
# Add the date and depth to the samples
AM_SAGs=pd.read_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/202108_GoM_resp_decon_1cell_SAGs.csv', index_col=0)

Tues=['AM-377', 'AM-384']
AM_SAGs['sample_collection_date']=np.where(AM_SAGs.Plate.isin(Tues), 210824, 210825)
AM_SAGs['sample_collection_time']=np.where(AM_SAGs.Plate.isin(Tues), 'Day', 'Night')

AM_SAGs['FACS date']='210928'

AM_SAGs['depth']=1

# add normalization values for RSG
cols=['date', 'slope (m)', 'intercept (b)']
RSG_conversion=pd.read_csv('531_conversion_factors.csv', usecols=cols)
RSG_conversion=RSG_conversion.rename(columns={'date':'FACS date'})
RSG_conversion['FACS date']=RSG_conversion['FACS date'].astype(str)
AM_SAGs=AM_SAGs.merge(RSG_conversion, on='FACS date', how='left')


AM_SAGs['normalized_531_value']=np.where(AM_SAGs['gate']=='Respiring cells (RSG-positive)', ((AM_SAGs['X531.40..488.']*AM_SAGs['slope (m)'])+AM_SAGs['intercept (b)']), '')

AM_SAGs.to_csv('/mnt/scgc/simon/microg2p/analyses/20210325_GoM_recluster/Summary_files/202108_GoM_resp_decon_1cell_SAGs_date_normalized_RSG.csv')
AM_SAGs

# The rows without a normalized 531 are all Syto9 stained SAGs and do not need to be converted.

```
